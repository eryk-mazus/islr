```python
import numpy as np
import pandas as pd 

from matplotlib import pyplot as plt
import seaborn as sns
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression

from scipy.special import expit

import statsmodels.api as sm
from patsy import dmatrices
```

```python
file_name = 'Default.xlsx'
df_de = pd.read_excel(f'data/{file_name}', index_col=0)
```

```python
df_de['default'].value_counts(normalize=True)
```

```python
df_de_subsample = pd.concat((df_de.loc[df_de['default'] == 'No'].sample(frac=.1), 
                            df_de.loc[df_de['default'] == 'Yes']))
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, gridspec_kw={'width_ratios': [2, 1, 1]}, figsize=(15,5))

fig.suptitle('The Default data set', fontsize=13)
sns.scatterplot(ax=ax1, data=df_de_subsample, x="balance", y="income", hue="default", style="default")
sns.boxplot(ax=ax2, x='default', y='balance', data=df_de)
sns.boxplot(ax=ax3, x='default', y='income', data=df_de)
```

### Linear and Logistic Regression comparison

```python
df_de['default_enc'] = df_de['default'].astype("category").cat.codes
```

```python
reg = LinearRegression().fit(df_de['balance'].values.reshape(-1,1), df_de['default_enc'])
print('Intercept:', reg.intercept_, '\nCoef:', reg.coef_)
```

```python
reg_log = LogisticRegression(solver='liblinear').fit(df_de['balance'].values.reshape(-1,1), df_de['default_enc'])
print('Intercept:', reg_log.intercept_, '\nCoef:', reg_log.coef_)
```

```python
X_test = np.linspace(df_de['balance'].values.min(), df_de['balance'].values.max(), 300)
loss = expit(X_test * reg_log.coef_ + reg_log.intercept_).ravel()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,4))
ax1.hlines([0,1], df_de['balance'].values.min()-150, df_de['balance'].values.max()+150, linestyle='dotted', color='black')
ax1.scatter(df_de['balance'], df_de['default_enc'], c='darkorange', edgecolors='black', linewidths=.5, alpha=.5)
ax1.plot(X_test, reg.coef_ * X_test + reg.intercept_, color='dodgerblue')
ax1.set_xlabel('Balance')
ax1.set_ylabel('Probability of Default')

ax2.hlines([0,1], df_de['balance'].values.min()-150, df_de['balance'].values.max()+150, linestyle='dotted', color='black')
ax2.scatter(df_de['balance'], df_de['default_enc'], c='darkorange', edgecolors='black', linewidths=.5, alpha=.5)
ax2.plot(X_test, loss, color='dodgerblue')
ax2.set_xlabel('Balance')
ax2.set_ylabel('Probability of Default')
```

### Multiple Logistic Regression

```python
df_de['student_enc'] = df_de['student'].astype("category").cat.codes
```

```python
y, X = dmatrices('default_enc ~ student_enc + balance + income', data=df_de, return_type='dataframe')
```

```python
mod = sm.Logit(y, X)
res = mod.fit()
print(res.summary())
```

The p-values associated with balance and the dummy variable for student status are very small, indicating that each of these variables is associated with the probability of default

### Linear Discriminant Analysis

#### One predictor (p=1) case

```python
# Generating the data - two normal density functions
mu1, mu2 = -1.25, 1.25
sigma = 1

s1 = np.random.normal(mu1, sigma, 20).reshape(-1,1)
s2 = np.random.normal(mu2, sigma, 20).reshape(-1,1)

s1 = np.append(s1, np.zeros(20).reshape(-1,1), axis=1)
s2 = np.append(s2, np.ones(20).reshape(-1,1), axis=1)

s = np.append(s1, s2, axis=0)
np.random.shuffle(s)

df = pd.DataFrame(data=s, columns=['x', 'y'])
```

```python
# Because prior probabilities are equal, Bayes decision boundary is equal to:
print(bdb := (mu1 + mu2)/2)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,4))
sns.distplot(ax=ax1, data=df, x="x", hue="y", kind="kde")
```
